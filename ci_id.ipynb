```python
import json

with open('./data/puid1.json', 'r') as file:
    data = json.load(file)

data
```

```python
ids = list(map(lambda x: x['id'], data['data']['allConfigurationItems']))
ids
```

```python
create_table = 'INSERT INTO puid (id) VALUES' + ', '.join(map(lambda x: f'({x})', ids))
with open('sql/create_puid_id.sql', 'w') as f:
    f.write(create_table)
```
